## Tüm telefonların özelliklerinin döngüyle çekilmesi

```python
len(total)
```

```python
telefon_title = []
telefon_content = []

for link in total:
    title = []
    content  = []
    
    # İlan sitesine bağlanıyoruz..
    page = requests.get(link, headers = headers)
    soup = BeautifulSoup(page.content, "html.parser")
    
    # Telefon markasını çekiyoruz..
    aziz = soup.find("span", attrs = {"class":"brand-name"})
    marka = aziz.text.replace("\n","")
    title.append("marka")
    content.append(marka)
    
    # Telefon modelini çekiyoruz..
    model = soup.find("h1", attrs = {"id":"product-name"}).text.replace("\n","").replace("\r","").replace("        ","").replace("    ","")
    title.append("model")
    content.append(model)
    
    # Telefonun fiyatını çekiyoruz..
    aziz2 = soup.find("span", attrs = {"data-bind":"markupText:'currentPriceBeforePoint'"})
    aziz3 = soup.find("span", attrs = {"data-bind":"markupText:'currentPriceAfterPoint'"})
    fiyat = aziz2.text.replace(".","") + "." + aziz3.text
    fiyat = float(fiyat)
    title.append("fiyat")
    content.append(fiyat)
    
    # Telefona ait diğer ek bilgileri çekiyoruz..
    aziz = soup.find_all("table", attrs = {"class":"data-list tech-spec"})
    feature = len(aziz[0].find_all("th"))
    for i in range(feature):
        adi = aziz[0].find_all("th")[i].text
        degeri = aziz[0].find_all("span")[i].text
        title.append(adi)
        content.append(degeri)
        
    
    telefon_title.append(title)
    telefon_content.append(content)
    print(link)
```

```python
frame = []
for i in range(len(telefon_title)):
    df = pd.DataFrame()
    df["title"] = telefon_title[i]
    df["content"] = telefon_content[i]
    
    myVars = vars()
    myVars[("df_"+str(i))] = df.transpose()
    myVars[("df_"+str(i))].columns = myVars[("df_"+str(i))].iloc[0]
    myVars[("df_"+str(i))] = myVars[("df_"+str(i))].iloc[1:]
    frame.append(myVars[("df_"+str(i))])
result = pd.concat(frame)
data = result
```
